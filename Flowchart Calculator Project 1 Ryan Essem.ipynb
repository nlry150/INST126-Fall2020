```python
# WELCOME TO THE 5-OPERATION CALCULATOR
# This program is a 5-OPERATION calculator that can add, subtract, multiply, divide AND calculate the powers of 'x'

# Define function adds two numbers

def add(x,y):
    return x + y

# Define function subtracts two numbers

def subtract(x,y):
    return x - y


# Define function multiplies two numbers

def multiply(x,y):
    return x * y


# Define function divides two numbers


def divide(x,y):
    return x / y


# Define function raises a number to a power of another number


def power(x,y):
    return x ** y


def calculate():
    op_sign = input('''
Please type in the math operation you would like to complete:
1. for Addition
2. for Subtraction
3. for Multiplication
4. for Division
5. for Power
''')
        
    num_1= float(input("Enter the first number: "))
    num_2= float(input("Enter the second number: "))

    if op_sign == '1':
        print('{} + {} = '.format(num_1, num_2))
        print(num_1 + num_2)

    elif op_sign == '2':
        print('{} - {} = '.format(num_1, num_2))
        print(num_1 - num_2)

    elif op_sign == '3':
        print('{} * {} = '.format(num_1, num_2))
        print(num_1 * num_2)

    elif op_sign == '4':
        print('{} / {} = '.format(num_1, num_2))
        print(num_1 / num_2)

    elif op_sign == '5':
        print('{} ** {} = '.format(num_1, num_2))
        print(num_1 ** num_2)

    else: 
        print('Invalid operation, please run program again using valid operations. ')
    
    again()
    
def again():
    calc_again = input('''Do you want to perform another calculation? Please type Y for YES or N for NO.''')

    if calc_again.upper() == 'Y':
        calculate()
    elif calc_again.upper() == 'N':
        print('Thank you for using this INST126 calculator program!')
    else:
        again()
calculate()


    
```
